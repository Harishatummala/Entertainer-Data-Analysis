```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
e1=pd.read_csv("/content/Entertainer - Basic Info.csv")
e1.head()
```

```python
e2=pd.read_csv("/content/Entertainer - Breakthrough Info.csv")
e2.head()
```

```python
e3=pd.read_csv("/content/Entertainer - Last work Info.csv")
e3.head()
```

```python
#Concatenating three tables for analysis and visualisation
df= pd.concat([e1,e2,e3], axis = 1 , join='outer')
Entertainer_df = df.loc[:,~df.columns.duplicated()].copy() # To remove the duplicate columns
Entertainer_df.head()
```

```python
#To find the age of entertainer at the time of receiving first award.
Entertainer_df['Age_at_first_award']=Entertainer_df['Year of First Oscar/Grammy/Emmy']-Entertainer_df['Birth Year']
Entertainer_df
```

3 Data sets are concatenated to form a single Dataset

```python
df = Entertainer_df
df['Age_at_first_award'] = df['Age_at_first_award'].fillna(0)

Entertainer_df['Age_at_first_award'] = df['Age_at_first_award'].astype(np.int64)
```

```python
plt.figure(figsize=(20, 18))
plt.bar(Entertainer_df['Entertainer'], Entertainer_df['Age_at_first_award'], color='skyblue')
plt.xlabel('Entertainers')
plt.ylabel('Age_at_first_award')
plt.title('Age at the time of winning awards')
plt.grid(axis='y', linestyle='--', alpha=0.5)
Entertainer_df.groupby("Entertainer").size().plot(kind="bar")
# Show the plot
plt.show()
```

```python
Entertainer_df = Entertainer_df[(Entertainer_df['Age_at_first_award'] > 0)]
top10=Entertainer_df['Age_at_first_award'].sort_values(ascending=True)[:10]

top10.plot(kind='bar')

plt.xlabel('Entertainers')
plt.ylabel('Age')
plt.title('Top 10 Youngest award winners')
plt.xticks(rotation=45)
plt.show()
for key in top10.keys():
  print(key,"-",Entertainer_df['Entertainer'].loc[key],"at",top10[key])
```

```python
top10=Entertainer_df['Age_at_first_award'].sort_values(ascending=False)[:10]

top10.plot(kind='bar')

plt.xlabel('Entertainers')
plt.ylabel('Age')
plt.title('Top 10 Oldest award winners')
plt.xticks(rotation=45)
plt.show()
for key in top10.keys():
 print(key,"-",Entertainer_df['Entertainer'].loc[key],"at",top10[key])
```

```python
#Total no of award winners
young_awardees=(Entertainer_df['Age_at_first_award']).count()
print(young_awardees)
```

```python
#No of Award winners below 30 years 0f age
count=0
for i in Entertainer_df['Age_at_first_award']:
  if i<=30:
    count+=1
print(count)
```

```python
#No of Award winners below 30 years 0f age
count=0
for i in Entertainer_df['Age_at_first_award']:
  if i>=60:
    count+=1
print(count)
```

```python
#Entertainers whose first breakthrough/nomination year is same as year of winning first award
Entertainer_df['Diff_Nom_award']=Entertainer_df['Year of First Oscar/Grammy/Emmy']-Entertainer_df['Year of Breakthrough/#1 Hit/Award Nomination']
count=0
for i in Entertainer_df['Diff_Nom_award']:
  if i==0:
    count+=1
print(count)
```
